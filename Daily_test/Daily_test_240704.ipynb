```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

ax1 = sns.scatterplot(x='total_bill', y='tip', data=tips, hue="time",ax=ax[0,0])

ax2 = sns.boxplot(data=tips, x="day", y="tip", hue="sex", ax=ax[0,1])

ax3 = sns.violinplot(data=tips, x="day", y="tip", hue="smoker", ax=ax[1,0])

ax4 = sns.histplot(data=tips, x="total_bill", ax=ax[1,1])


plt.tight_layout() #  
plt.savefig('yourname.png', dpi=300) #  
plt.show()
```
