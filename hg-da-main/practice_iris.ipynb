# 다중조건
- AND 연산자: 교집합, &
- OR 연산자: 합집합, |

```python
cols = ['sepal_width', 'petal_length']
result1 = iris.loc[iris['species'] == 'setosa']
result2 = result1.loc[iris['sepal_length']<=5.1]
result2.shape
```

```python
result = iris.loc[(iris['species'] == 'setosa') & (iris['sepal_length']<=5.1), cols].reset_index(drop=True)
result.shape
```
